```python
import pandas as pd

def read_and_add_filename(file_path):
    try:
        df = pd.read_csv(file_path)
        df['Review_Type'] = file_path[:-4]  # Extract filename from path
        return df
    except FileNotFoundError as e:
        print(f"Error reading {file_path}: {e}")
        return pd.DataFrame()  # Return an empty DataFrame for failed files

# List of CSV files (adjust paths accordingly)
file_paths = ['Deleted_Reviews.csv', 'Not_Published_Reviews.csv','Published_Reviews.csv'] 

# Read and process each file
dfs = []
for file_path in file_paths:
    df = read_and_add_filename(file_path)
    if not df.empty: 
        dfs.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df)
combined_df.to_csv('All_reviews.csv',index=False)
```

```python
import pandas as pd

# Sample DataFrame
data = {'product': ['A', 'A', 'A', 'B', 'B', 'C', 'C'], 
        'review_date': ['2023-11-01', '2023-11-15', '2023-12-05', '2023-11-10', '2023-12-10', '2023-12-01', '2023-12-15'], 
        'rating': [4, 5, 3, 2, 4, 5, 1],
        'revenue': [100, 150, 80, 50, 120, 200, 100]} 
df = pd.DataFrame(data)

# Convert 'review_date' to datetime
df['review_date'] = pd.to_datetime(df['review_date']) 

# Truncate review_date to month level
df['month'] = df['review_date'].dt.to_period('M')

# Pivot the DataFrame
pivot_table = df.pivot_table(
    index='product', 
    columns='month', 
    values=['rating', 'revenue'], 
    aggfunc={'rating': ['count', 'mean'], 'revenue': 'sum'}
)

print(pivot_table)
```
